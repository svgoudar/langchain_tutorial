```python
# cell 1: Define your RAG logic and FastMCP Server
from fastmcp import FastMCP
from llama_index.core import VectorStoreIndex, Document

mcp = FastMCP("RAG-Server")

# Dummy RAG setup for testing
docs = [Document(text="The secret code is 12345."), Document(text="Project Alpha ends in Dec 2025.")]
index = VectorStoreIndex.from_documents(docs)
query_engine = index.as_query_engine()

@mcp.tool()
def query_knowledge_base(question: str) -> str:
    """Queries the internal RAG system for answers."""
    response = query_engine.query(question)
    return str(response)

# cell 2: Test the tool function directly (Unit Testing)
# This bypasses MCP protocol to ensure your RAG logic is sound
print(query_knowledge_base("What is the secret code?"))
```
