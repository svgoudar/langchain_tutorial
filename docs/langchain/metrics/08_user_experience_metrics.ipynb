---

### Task Resolution Rate

```python
completed = 87
total = 100
resolution_rate = completed / total
print("Resolution Rate:", resolution_rate)
```

---

### Escalation Rate

```python
escalations = 12
escalation_rate = escalations / total
print("Escalation Rate:", escalation_rate)
```

---

### Conversation Length

```python
convo_turns = [3, 4, 2, 5, 6]
avg_turns = sum(convo_turns) / len(convo_turns)
print("Avg Conversation Length:", avg_turns)
```

---

### Abandonment Rate

```python
abandoned = 9
abandonment_rate = abandoned / total
print("Abandonment Rate:", abandonment_rate)
```

---

### Repeat Usage Rate

```python
returning_users = 42
active_users = 100
repeat_usage_rate = returning_users / active_users
print("Repeat Usage Rate:", repeat_usage_rate)
```

---

### How UX Metrics Drive Decisions

| Problem         | Revealed By         |
| --------------- | ------------------- |
| Low trust       | CSAT                |
| Poor automation | Escalation rate     |
| Confusing flows | Conversation length |
| Bad onboarding  | Abandonment rate    |
| Low retention   | Repeat usage        |

---

### Typical Production Targets

| Metric           | Target    |
| ---------------- | --------- |
| CSAT             | ≥ 4.2 / 5 |
| Resolution Rate  | ≥ 85%     |
| Escalation Rate  | ≤ 10%     |
| Abandonment Rate | ≤ 5%      |
| Repeat Usage     | ≥ 40%     |
